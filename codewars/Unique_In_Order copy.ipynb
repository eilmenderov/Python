```python
# Implement the function unique_in_order which takes as argument a sequence and
# returns a list of items without any elements with the same value next to each
# other and preserving the original order of elements.

def unique_in_order(iterable):
    str_len = len(iterable)
    if str_len == 0:
        return ([])
    rez = []
    rez.append(iterable[0])
    char_last = iterable[0]
    i = 0
    while (i < str_len):
        if (char_last != iterable[i]):
            rez.append(iterable[i])
            char_last = iterable[i]
        i += 1
    return (rez)

string = input()
print(unique_in_order(string))
```
