```python
# The new "Avengers" movie has just been released!
# There are a lot of people at the cinema box office standing in a huge line. 
# Each of them has a single 100, 50 or 25 dollar bill. An "Avengers" ticket costs 25 dollars.
# Vasya is currently working as a clerk. He wants to sell a ticket
# to every single person in this line. Can Vasya sell a ticket to every person
# and give change if he initially has no money and sells the tickets strictly in the order people queue?
# Return YES, if Vasya can sell a ticket to every person and give change
# with the bills he has at hand at that moment. Otherwise return NO.

def tickets(people):
    m_25 = people.count(25)
    m_50 = people.count(50)
    m_100 = people.count(100)
    if (m_25 < m_50):
        return "NO"
    m_25 = m_25 - m_50
    min_c = m_25
    if (m_25 > m_50):
        min_c = m_50
    m_100 = m_100 - min_c
    if (m_100 <= 0):
        return "YES"
    m_25 = m_25 - min_c
    if (m_25 // 3 >= m_100):
        return "YES0"
    return "NO"
    

people = list(map(int, input().split()))
print(people)
print(tickets(people))
```
