```python
# In this little assignment you are given a string of space separated numbers,
# and have to return the highest and lowest number.

def high_and_low(numbers):
    num = numbers.split(' ')
    min_num = int(num[0])
    max_num = int(num[0])
    i = 0
    while (i < len(num)):
        if (min_num > int(num[i])):
            min_num = int(num[i])
        if (max_num < int(num[i])):
            max_num = int(num[i])
        i += 1
    numbers = ""
    return (numbers + str(max_num) + " " + str(min_num))

num = input()
print(high_and_low(num))
```
