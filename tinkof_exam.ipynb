```python
# Бабушка подарила Лизе две коробки с конфетами: в первой коробке ﻿AAA﻿ конфет, а во второй — ﻿BBB﻿ конфет. Мама Лизы не разрешает ей есть больше, чем ﻿NNN﻿ конфет в день. Сколько конфет сегодня сможет съесть Лиза (при условии, что ей хочется съесть как можно больше конфет)?
# Формат входных данных
# В трех строках вводятся три целых положительных числа: ﻿AAA﻿, ﻿BBB﻿, ﻿NNN﻿ (﻿1≤A,B,N≤10001 \le A, B, N \le 10001≤A,B,N≤1000﻿).
# Формат выходных данных
# Выведите одно число — максимальное количество конфет, которое может съесть Лиза.

a = int(input())
b = int(input())
n = int(input())
if (a + b < n):
    print(a + b)
else:
    print(n)
```

```python
# Вы пришли в магазин, в котором находится ﻿nnn﻿ скатертей. Вам нужно купить самую дешевую скатерть для вашего прямоугольного стола размера ﻿W×HW \times HW×H﻿. У каждой скатерти есть свои габариты ﻿wi×hiw_i \times h_iwi​×hi​﻿, а также своя стоимость в ﻿cic_ici​﻿ единиц.
# Считается, что скатерть покрывает стол, если верхнюю поверхность стола не видно под скатертью. Найдите самую дешевую скатерть, покрывающую стол.
# Формат входных данных
# В первой строке вводится натуральное число ﻿nnn﻿ ﻿(1≤n≤1000)(1 \le n \le 1000)(1≤n≤1000)﻿﻿ −~- −﻿ количество скатертей в магазине.
# Во второй строке вводятся два натуральных числа ﻿WWW﻿, ﻿HHH﻿ ﻿(1≤W, H≤1000)(1 \le W,\,H \le 1000)(1≤W,H≤1000)﻿﻿ −~- −﻿ размеры вашего стола.
# Во следующих ﻿nnn﻿ строках вводятся натуральные числа ﻿wi, hi, ciw_i,\,h_i,\,c_iwi​,hi​,ci​﻿ ﻿(1≤wi, hi, ci≤1000)(1 \le w_i,\,h_i,\,c_i \le 1000)(1≤wi​,hi​,ci​≤1000)﻿﻿ −~- −﻿габариты скатерти и ее стоимость.
# Формат выходных данных
# Выведите одно число — оптимальную стоимость покупки скатерти.

n = int(input())
w_h = list(map(int, input().split()))
rez = 0;
for i in range(n):
    whc = list(map(int, input().split()))
    if ((whc[0] >= w_h[0] and whc[1] >= w_h[1]) or (whc[1] >= w_h[0] and whc[0] >= w_h[1])):
        if (rez == 0):
            rez = whc[2]
        elif (whc[2] < rez):
            rez = whc[2]
print(rez)
```

```python
# Вам дано очень большое число ﻿NNN﻿. В этом числе один символ скрыт — на его месте стоит звездочка.
# От вас требуется сказать, на какие числа от ﻿1﻿ до ﻿10﻿ число ﻿N﻿ делится без остатка гарантированно (то есть вне зависимости от того, какая цифра скрыта за звездочкой)?
# Формат входных данных
# Вводится единственная строка, обозначающая число ﻿N﻿ ﻿(10<N<10**1000)(10 < N < 10 ^{1000} )(10<N<10**1000)﻿. В строке содержится ровно один символ ﻿«∗»«*»«∗»﻿.
# Формат выходных данных
# Выведите все числа от ﻿111﻿ до ﻿101010﻿, на которые гарантированно делится ﻿NNN﻿. Числа выводить следует по возрастанию через пробел.

def create_str(st_r):
    rez = '1'
    l_str = len(st_r)
    if (st_r[l_str - 1] == '*'):
        return (rez)
    if int(st_r[l_str - 1]) % 2 == 0:
        rez = rez +' 2'
    if (l_str >= 3 and st_r[l_str - 2] != '*' and int(st_r[l_str - 2:l_str - 1]) % 4 == 0):
        rez = rez + ' 4'
    if (st_r[l_str - 1]) == '0' or (st_r[l_str - 1]) == '5':
        rez = rez +' 5'
    if (l_str >= 4 and st_r[l_str - 2] != '*' and st_r[l_str - 3] != '*' and int(st_r[l_str - 3:l_str - 1]) % 8 == 0):
        rez = rez + ' 8'
    if (st_r[l_str - 1]) == '0':
        rez = rez +' 10'
    return (rez)
    

st_r = input()
rez = create_str(st_r)
print(rez)
```

```python
# У Петра Васильевича есть ﻿nnn﻿ банковских счетов. Он устал следить за всеми карточками и решил наконец-то перевести деньги на один счет.
# Процесс, проводимый Петром Васильевичем, выглядит так: он берет один из своих банковских счетов, переводит все деньги с этого счета на другой, затем закрывает пустой счет. Процесс закончится, когда у Петра Васильевича останется ровно один счет.
# При переводе ﻿ААА﻿ рублей с одного счета на другой счет с ﻿ВВВ﻿ рублями Петр Васильевич платит банку комиссию ﻿5% 5\%5%﻿ от суммы ﻿A+BA + BA+B﻿.
# Помогите Петру Васильевичу определить минимальную сумму, которую придется потратить на комиссию.
# Формат входных данных
# В первой строке вводится число ﻿nnn﻿ ﻿(2≤n≤105)(2 \le n \le 10^5)(2≤n≤105)﻿﻿ −~- −﻿ количество банковских счетов.
# В следующей строке вводится ﻿nnn﻿ чисел﻿ −~- −﻿ суммы, лежащие на банковских счетах Петра Васильевича. Каждое число ﻿−-−﻿ натуральное, не превосходящее ﻿100001000010000﻿.
# Формат выходных данных
# Определите минимальную сумму, которая потратится на комиссию.

n = int(input())
mass = list(map(int, input().split()))
rez =  0.0;
while (len(mass) > 1):
    min_mas1 = min(mass)
    mass.remove(min_mas1)
    min_mas2 = min(mass)
    mass.remove(min_mas2)
    mass.append(min_mas1 + min_mas2)
    rez = rez + (min_mas1 + min_mas2) * 0.05
print("%.2f" % rez)
```

```python
# Вам дан многочлен ﻿f(x)=a0+a1⋅x+a2⋅x2+...+an⋅xnf (x) = a_0 + a_1 · x + a_2 · x ^2 + . . . + a_n · x^nf(x)=a0​+a1​⋅x+a2​⋅x2+...+an​⋅xn﻿ . Гарантируется, что ﻿nnn﻿ нечетное. Найдите такой ﻿xxx﻿, что ﻿f(x)=0f(x) = 0f(x)=0﻿. Гарантируется, что все такие ﻿xxx﻿ лежат в диапазоне ﻿[−100:100][−100 : 100][−100:100]﻿.
# Формат входных данных
# В первой строке вводится натуральное число ﻿nnn﻿ (﻿1≤n≤5, n=2k−11 \le n \le 5,\ n = 2k - 11≤n≤5, n=2k−1﻿, где ﻿kkk﻿ натуральное).
# В следующей строке вводится ﻿n+1n + 1n+1﻿ вещественное число ﻿a0, a1,…, ana_0,\,a_1,\ldots,\,a_na0​,a1​,…,an​﻿ ﻿(ai∈[−10000; −0.0001] ∪ {0} ∪[0.0001; 10000])(a_i \in [-10000;\,-0.0001]\ \cup\ \{0\}\ \cup [0.0001;\,10000])(ai​∈[−10000;−0.0001] ∪ {0} ∪[0.0001;10000])﻿. 
# Гарантируется, что ﻿an>0a_n > 0an​>0﻿.
# Формат выходных данных
# Выведите одно число — найденный корень многочлена. Ваш ответ будет считаться верным, если значение многочлена в этой точке не превосходит по абсолютной величине ﻿10−610^{-6}10−6﻿. 

def power(x, i):
    r = 1
    while i:
        if i & 1:
            r *= x
        x *= x
        i >>= 1
    return r

def f_x_ravno(n, a_arr, x):
    rez = 0.0
    i = 0
    while (i < n + 1):
        rez = rez + a_arr[i]*(power(x, i))
        i += 1
    return (rez)

n = int(input())
a_arr = list(map(float, input().split()))
x = -100.0
while (x <= 100):
    if (f_x_ravno(n, a_arr, x) == 0):
        break
    x += 0.0001
print(x)
```
