```python
# Task 1 Задан словарь. Напишите программу, которая будет выводить значение по заданному ключу.
n = int(input('Введите длину словаря: '))
d = dict()
for i in range(n):
    d.update({input('Введите ключ: '):input('Введите значение: ')})
key = input('Введите ключ для поиска: ')
print("rez =", d.get(key))
```

```python
# Task 2 Напишите программу, которая будет выполнять действие, обратное заданию 1.
# Программа должна производить поиск по значению и выдавать ключ.
n = int(input('Введите длину словаря: '))
d = dict()
for i in range(n):
    d.update({input('Введите ключ: '):input('Введите значение: ')})
d_val = list(d.values())
d_keys = list(d.keys())
k_len = len(d_keys)
var = input('Введите значение для поиска: ')
i = 0
while (i < k_len):
  if (d_val[i] == var):
    break
  i += 1
print('rez =', d_keys[i])
```
