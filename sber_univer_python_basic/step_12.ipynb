```python
# Task 1 Дан список чисел. 
# Напишите программу, которая определяет, сколько в нем встречается различных чисел, используя множества.
# Входные данные:    1, 2, 3, 2, 1
# Выходные значения: 3
n = int(input('Введите кол-во чисел: '))
s = set()
for i in range(n):
    s.update({input('Введите число № ' + str(i+1) + ': ')})
list_a = s
print('rez :', len(set(list_a)))
```

```python
# Task 2 Даны два списка чисел. 
# Напишите программу, которая определяет, сколько в них встречается общих чисел, используя множества.
# Входные данные:     1, 2, 3
#                     1, 4, 5
# Выходные значения:  1
n_1 = int(input('Введите кол-во чисел первого списка: '))
n_2 = int(input('Введите кол-во чисел второго списка: '))
s_1 = set()
s_2 = set()
for i in range(n_1):
    s_1.update({input('Введите число № ' + str(i+1) + ' для первого списка: ')})
for i in range(n_2):
    s_2.update({input('Введите число № ' + str(i+1) + ' для второго списка: ')})
list_a = s_1
list_b = s_2
man_a = set(list_a)
man_b = set(list_b)
man_rez = man_a.intersection(man_b)
print("rez :", len(man_rez))
```

```python
# Task 3 Даны два множества чисел. 
# Напишите программу, которая определяет, является ли первое множество подмножеством второго. 
# Множество является подмножеством другого множества, если все данные первого совпадают с частью данных из второго.
# Если множества совпадают, они не являются подмножествами друг друга.
# Входные данные:     {1, 2, 3}
#                     {1, 4, 5}
# Выходные значения:  False
n_1 = int(input('Введите кол-во чисел первого множества: '))
n_2 = int(input('Введите кол-во чисел второго множества: '))
s_1 = set()
s_2 = set()
for i in range(n_1):
    s_1.update({input('Введите число № ' + str(i+1) + ' для первого множества: ')})
for i in range(n_2):
    s_2.update({input('Введите число № ' + str(i+1) + ' для второго множества: ')})
man_a = s_1
man_b = s_2
rez = man_a.issubset(man_b)
print(rez)
```
