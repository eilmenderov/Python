```python
# Task 2
# Напишите код, который обрабатывает строковые данные и возвращает их с первыми
# заглавными буквами в каждом слове.
str =  'can    You   solve  it?'
tmp = str.split()
str = str.title()
rez = ""
i = 0
while (i < len(tmp)):
    if ((tmp[i][0]).isalpha() == True):
        tmp[i] = tmp[i].title()
    i += 1
i = 0
while (i < len(tmp)):
    rez = rez + tmp[i]
    i += 1
    if (i != len(tmp)):
        rez = rez + " "
print(rez)
```

```python
# Task 3
# Напишите код, который обрабатывает строковые данные и выводит сообщение о том,
# надежен ли пароль или нет. В случае, еслипароль не надежен, код должен выдавать
# рекомендации для усиления надежности пароля.

password = input('Введите пароль: ')
digits = '1234567890'
digits_count = 0
leters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
leters_count = 0
leters_2 = 'abcdefghijklmnopqrstuvwxyz'
leters_count_2 = 0
symbols = '!@#$%^&*()-+'
symbols_count = 0
for i in password:
    if i in digits:
        digits_count+=1
    elif i in leters:
        leters_count+=1
    elif i in leters_2:
        leters_count_2+=1
    elif i in symbols:
        symbols_count+=1
    else:
        print('Ошибка. Запрещенный спецсимвол: '+i)
s = ''
if digits_count + leters_count + leters_count_2 + symbols_count < 12:
    s+= "увеличить число символов -"+str(12-digits_count + leters_count + leters_count_2 + symbols_count)
if digits_count < 1:
    if s[-1:]!='':
        s+=', '
    s+= "добавить 1 цифру"
if leters_count < 1:
    print(s[-1:])
    if s[-1:]!='':
        s+=', '
    s+= "добавить 1 заглавную букву"
if leters_count_2 < 1:
    if s[-1:]!='':
        s+=', '
    s+= "добавить 1 строчную букву"
if symbols_count < 1:
    if s[-1:]!='':
        s+=', '
    s+= "добавить 1 спецсимвол"
if s == '':
    print('Сильный пароль')
else:
    print('Слабый пароль. Рекомендации:'+s)
```
