```python
n = 5
i = 1
while (i <= n):
    j = 1
    while (j <= i):
        print(j, end = '')
        j += 1
    print("\n")
    i += 1
```

```python
n = 5
i = 1
tmp = n
while (i <= n and tmp >= 0):
    j = 1
    while (j != tmp):
        print(" ", end = "")
        j += 1
    j = 1
    while (j <= i):
        print(j, end = '')
        j += 1
    j -= 1
    while (j >= 1):
        if (j != i):
            print(j, end = '')
        j -= 1
    tmp -= 1
    i += 1
    print("\n")
```
