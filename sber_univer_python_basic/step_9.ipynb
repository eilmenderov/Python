```python
# Task 1 Задан список с числами. 
# Напишите программу, которая выводит все элементы списка с четными индексами в виде нового списка.
a = ['1', '2', '3', '4', '5', '6']
rez = []
len_a = len(a)
i = 0
while (i < len_a):
    if (i % 2 == 0):
        rez.append(a[i])
    i += 1
print(rez)
```

```python
# Task 2 Задан список с числами.
# Напишите программу, которая выводит все элементы списка, которые больше предыдущего, в виде отдельного списка.
a = ['1', '5', '2', '4', '3']
rez = []
len_a = len(a)
i = 1
while (i < len_a):
    if (a[i] > a[i - 1]):
        rez.append(a[i])
    i += 1
print(rez)
```

```python
# Task 3 Задан список с числами.
# Напишите программу, которая меняет местами наибольший и наименьший элемент и выводит новый список.
a = ['1', '5', '2', '4', '9']
len_a = len(a)
min_a = a[0]
max_a = a[0]
i = 1
while (i < len_a):
    if (max_a < a[i]):
        max_a = a[i]
    if (min_a > a[i]):
        min_a = a[i]
    i += 1
ind_max = a.index(max_a)
ind_min = a.index(min_a)
a.pop(ind_max)
a.insert(ind_max, min_a)
a.remove(min_a)
a.insert(ind_min, max_a)
print(a)
```
