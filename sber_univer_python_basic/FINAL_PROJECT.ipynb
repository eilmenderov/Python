```python
import random
def pass_gen(k):
    up_sym = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    low_sym = 'abcdefghijklmnopqrstuvwxyz'
    num_sym = '1234567890'
    spec_sym = '!@#$%^&*()-+'
    dict_for_password = up_sym + num_sym + low_sym + spec_sym
    rez = []
    rez.append(random.choice(up_sym))
    rez.append(random.choice(num_sym))
    rez.append(random.choice(low_sym))
    rez.append(random.choice(spec_sym))
    while(len(rez) < k):
      rez.append(random.choice(dict_for_password))
    random.shuffle(rez)
    pas = ""
    i = 0
    while(i < len(rez)):
      pas += rez[i]
      i += 1
    return(pas)

def email_gen(list_of_names):
    emails = []
    for i in list_of_names:
        letter = 1
        while i[1] + '.' + i[0][0:letter] + '@company.io' in emails:
            letter+=1
        emails.append(i[1] + '.' + i[0][0:letter] + '@company.io')
    return emails

f = open('task_file.txt')
tmp = f.read()
f.close()
tmp = tmp.split('\n')
kkk = "PASSWORD, " + tmp[0] + '\n'
rez = []
answ = []
i = 1
while (i < len(tmp) - 1):
    key = tmp[i]
    key = key.replace(',', '')
    if (key[0] == ' '):
        key = key[1:]
    key = key.split(' ')
    if ((key[0] != '' and key[1] != '' and key[2] != '' and key[3] != '')
        and key[0] != key[1] and key[0].istitle() == True and key[1].istitle() == True
        and len(key[2]) == 7 and key[2].isdigit() == True):
        answ.append(key)
        key = key[:2]
        rez.append(key)
    i += 1
rez = email_gen(rez)
i = 0
while (i < len(rez)):
    answ[i].insert(0, rez[i])
    answ[i].insert(0, pass_gen(12))
    i += 1
i = 0
while (i < len(answ)):
    j = 0
    while (j < 6):
        kkk += answ[i][j]
        if (j != 4):
            kkk += ', '
        else:
            kkk += ','
        j += 1
    kkk += '\n'
    i += 1
fd = open('rez.txt', 'w')
fd.write(kkk)
fd.close()
```
